```python
pf=pd.read_excel(r'C:\Users\LENOVO\Downloads\output.xlsx', engine='openpyxl')
pf.head()
```

```python
y = pf[['TAH ']]
```

```python
x=df
```

```python
#importing the sklearn library for train test data split
from sklearn import preprocessing
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
```

```python


x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y,test_size=0.30)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

#now prediction of the model
y_pred=lr.predict(x_test)




#now for the sake of comparison we are using r2 method
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
